## 请删除最后一列为缺失值的行

```python
#请删除最后一列为缺失值的行
zz = zz[zz['发货地址'].notna()]
zz

#查看异常的价格
zz[~zz['价格'].str.contains(r'^-?(\d+)(\.\d+)?$')]
zz = zz[zz['价格'].str.contains(r'^-?(\d+)(\.\d+)?$')]
#求所有在杭州发货的商品单价均值
zz[zz['发货地址'].str.contains('杭州')]['价格'].astype('float').mean()
```

## 商品标题带有“嘉兴”但发货地却不在嘉兴的商品有多少条记录？

```python
zz[zz['标题'].str.contains('嘉兴') & ~zz['发货地址'].str.contains('嘉兴')]
```

## 请按照分位数将价格分为“高、较高、中、较低、低”5 个类别，再将类别结果插入到标题一列之后，最后对类别列进行降序排序

```python
zz['价格'] = zz['价格'].astype('float')
```

```python
zz['类别'] = pd.qcut(zz['价格'],[0,0.2,0.4,0.6,0.8,1.],labels=['低','较低','中','较高','高'])
zz.sort_values('类别', ascending=False)
```

```python
zz[zz.付款人数.isna()]
```

```python
df['year'] = df.Date.apply(lambda x:x.year)
df['month'] = df.Date.apply(lambda x:x.month)
df['day'] = df.Date.apply(lambda x:x.day)
df['weekday'] = df.Date.apply(lambda x:x.weekday())
```
